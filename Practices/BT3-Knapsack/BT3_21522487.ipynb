```python
from ortools.algorithms.python import knapsack_solver
from matplotlib import pyplot as plt
import os
import numpy as np
import sys
import time
```

```python
solver = knapsack_solver.KnapsackSolver(
    knapsack_solver.SolverType.KNAPSACK_MULTIDIMENSION_BRANCH_AND_BOUND_SOLVER, 'KnapsackExample')
```

```python
def knapsack(data, timelimit):
    solver.init(data["values"],data["weights"],data["capacities"])

    solver.set_time_limit(timelimit)
    time_start = time.time()
    computed_value = solver.solve()
    time_end = time.time()

    runtime = time_end - time_start
    isOptimal = (runtime <= timelimit)

    packed_items = []
    packed_weights = []
    total_weight = 0
    for i in range(len(data["values"])):
        if solver.best_solution_contains(i):
            packed_items.append(i)
            packed_weights.append(data["weights"][0][i])
            total_weight += data["weights"][0][i]

    return {"packed_items":packed_items,
            "packed_weights":packed_weights,
            "total_weight":total_weight,
            "total_value":computed_value,
            "isOptimal":isOptimal,
            "runtime":runtime
            }
```

```python
root_path = "/content/drive/MyDrive/UIT/CS106-AI/BaiTap3-Knapsack"
groupTestCases = os.listdir(r'{}/kplib'.format(root_path))
groupSizeItem = os.listdir(r'{}'.format(root_path+'/kplib/'+groupTestCases[0]))
groupSizeR = ['R01000','R10000']
items = ['s000']
```

```python
groupTestCases
```

```python
for groupCase in groupTestCases:
    for sizeItem in [sizeItem for index, sizeItem in enumerate(groupSizeItem) if index < 6]:
        for sizeR in groupSizeR:
          for item in items:
              path_data = root_path + '/kplib/{}/{}/{}/{}.kp'.format(groupCase,sizeItem,sizeR,item)
              path_result = root_path + '/Result/{}/{}_{}_{}.txt'.format(groupCase,sizeItem,sizeR, item)
              path_res = root_path + '/Result/{}'.format(groupCase)
              data = {
                  "values":[],
                  "weights":[[]],
                  "capacities":[],
                  "size":[]
              }
              with open(path_data,'r') as file_data:
                  file = file_data.read().splitlines()
                  data["size"].append(int(file[1]))
                  data["capacities"].append(int(file[2]))
                  for i in range(4,len(file)):
                      data["values"].append(int(file[i].split(' ')[0]))
                      data["weights"][0].append(int(file[i].split(' ')[1]))

              result = knapsack(data,180)
              if not os.path.exists(path_res):
                  os.makedirs(path_res)
              with open(path_result, 'w+') as file_result:
                  file_result.write(f"Total value: {result['total_value']}\n")
                  file_result.write(f"Total weight: {result['total_weight']}\n")
                  file_result.write(f"Runtime: {result['runtime']:.6f}\n" )
                  file_result.write(f"Optimal: {result['isOptimal']}\n" )
                  file_result.write(f"Packed items: {result['packed_items']}\n")
                  file_result.write(f"Packed weights: {result['packed_weights']}\n")
```
